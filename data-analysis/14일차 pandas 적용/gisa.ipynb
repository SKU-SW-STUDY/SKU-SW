```python
# 기사문제1번 
gisa1 = pd.read_csv('Abc1115.csv', header=None, sep=',') # header=None : 컬럼을 기본 컬럼으로 지정
gisa1.columns = ['stdNo', 'email', 'kor', 'eng', 'math', 'sci', 'hist', 'total', 'mgrCode', 'accCode', 'locCode']
gisa1 = gisa1[gisa1['locCode'] == 'B']
```

```python
# 기사문제3번
gisa3 = pd.read_csv('Abc1115.csv', header=None, sep=',') # header=None : 컬럼을 기본 컬럼으로 지정
gisa3.columns = ['stdNo', 'email', 'kor', 'eng', 'math', 'sci', 'hist', 'total', 'mgrCode', 'accCode', 'locCode']
gisa3 = gisa3[gisa3['eng'] + gisa3['math'] >= 120]
```

```python
gisa3
```

```python
sum_result = 0

for index in range(len(gisa3)):
    row = gisa3.iloc[index]
    sum_result += pd.to_numeric(row['total'])
    
    if row['accCode'] == 'A':
        sum_result += 5
    elif row['accCode'] == 'B':
        sum_result += 15
    else:
        sum_result += 20

print(sum_result)

```

```python
# 기사문제4번
gisa4 = pd.read_csv('Abc1115.csv', header=None, sep=',') # header=None : 컬럼을 기본 컬럼으로 지정
gisa4.columns = ['stdNo', 'email', 'kor', 'eng', 'math', 'sci', 'hist', 'total', 'mgrCode', 'accCode', 'locCode']
gisa4 = gisa4[(gisa4['accCode'] == 'A') | (gisa4['accCode'] == 'B')]
```

```python
# point 컬럼을 만들어 계산하기 
gisa4['point'] = gisa4['locCode'].map({'A':5, 'B':10, 'C':15})
```
